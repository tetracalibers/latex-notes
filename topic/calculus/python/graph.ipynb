```python
import numpy as np
import matplotlib.pyplot as plt

# グリッド生成
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 - Y**2  # 関数 z = x^2 - y^2

# 描画設定
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# 3D曲面プロット
ax.plot_surface(X, Y, Z, cmap="rainbow", alpha=0.6, rstride=5, cstride=5)

# 各軸への等高線の投影
ax.contour(X, Y, Z, zdir='z', offset=-25, cmap='rainbow')
ax.contour(X, Y, Z, zdir='x', offset=-6, cmap='rainbow')
ax.contour(X, Y, Z, zdir='y', offset=6, cmap='rainbow')

# 軸範囲とラベル
ax.set(xlim=(-6, 6), ylim=(-6, 6), zlim=(-25, 25))

# Transparent spines
ax.xaxis.line.set_color((1.0, 1.0, 1.0, 0.0))
ax.yaxis.line.set_color((1.0, 1.0, 1.0, 0.0))
ax.zaxis.line.set_color((1.0, 1.0, 1.0, 0.0))

# No ticks
ax.set_xticks([]) 
ax.set_yticks([]) 
ax.set_zticks([])

ax.set_xlabel('X', labelpad=-11, fontsize=12, fontname="PilGi", color="lightslategray")
ax.set_ylabel('Y', labelpad=-12, fontsize=12, fontname="PilGi", color="lightslategray")
ax.set_zlabel('Z', labelpad=-14, fontsize=12, fontname="PilGi", color="lightslategray")

plt.savefig('graph_x-pow2-sub-y-pow2_01.png', dpi=300, bbox_inches="tight")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# グリッド生成
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 - Y**2  # 関数 z = x^2 - y^2

# 描画設定
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# 3D曲面プロット
ax.plot_surface(X, Y, Z, cmap="cool", alpha=0.4, rstride=5, cstride=5)

# z = x^2 のプロット
X2 = np.linspace(-5, 5, 100)
Z2 = X2**2
ax.plot(X2, np.zeros_like(X2), Z2, color='deeppink', label='$z = x^2$')

# z = -y^2 のプロット
Y2 = np.linspace(-5, 5, 100)
Z2 = -Y2**2
ax.plot(np.zeros_like(Y2), Y2, Z2, color='deepskyblue', label='$z = -y^2$')

# 凡例
ax.legend(loc='upper left', fontsize=10, framealpha=0.5)

# 軸範囲とラベル
ax.set(xlim=(-6, 6), ylim=(-6, 6), zlim=(-25, 25))

# Transparent spines
ax.xaxis.line.set_color((1.0, 1.0, 1.0, 0.0))
ax.yaxis.line.set_color((1.0, 1.0, 1.0, 0.0))
ax.zaxis.line.set_color((1.0, 1.0, 1.0, 0.0))

# No ticks
ax.set_xticks([]) 
ax.set_yticks([]) 
ax.set_zticks([])

ax.set_xlabel('X', labelpad=-11, fontsize=12, fontname="PilGi", color="lightslategray")
ax.set_ylabel('Y', labelpad=-12, fontsize=12, fontname="PilGi", color="lightslategray")
ax.set_zlabel('Z', labelpad=-14, fontsize=12, fontname="PilGi", color="lightslategray")

plt.savefig('graph_x-pow2-sub-y-pow2_02.png', dpi=300, bbox_inches="tight")
plt.show()
```
