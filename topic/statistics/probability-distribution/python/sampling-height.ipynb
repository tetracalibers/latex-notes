```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib_fontja # 日本語の文字化け防止
```

```python
minmax = (150, 200)

def calc_bins(interval):
    """階級幅に基づいてビンの範囲を計算する"""
    return np.arange(minmax[0], minmax[1] + interval, interval)

def generate_data(sample_size):
    """指定された標本数の身長データを生成する"""
    # ref: https://qiita.com/sakabe/items/df7fac64e6ffbb50848c
    return np.random.normal(171.7, 6.6, sample_size)

def draw_histogram(data, ax, bins):
    """相対度数を縦軸とするヒストグラムを描画する"""
    weights = np.ones_like(data) / len(data)
    ax.hist(data, bins=bins, weights=weights, color="tab:cyan", edgecolor="tab:gray", alpha=0.7)
    ax.set_xlabel('身長 (cm)')
    ax.set_ylabel('相対度数')
    
def draw_line_chart(data, ax, bins):
    """線グラフを描画する"""
    counts, _ = np.histogram(data, bins=bins)
    mid_points = (bins[:-1] + bins[1:]) / 2
    ax.plot(mid_points, counts / len(data), color="tab:orange", label="確率分布")
    ax.legend()

def f(x):
    """正規分布の確率密度関数"""
    return (1 / (6.6 * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - 171.7) / 6.6) ** 2)

def draw_graph(ax, interval, label="確率分布"):
    """グラフを描画する"""
    x = np.arange(minmax[0], minmax[1] + interval, 0.1)
    y = f(x)
    ax.plot(x, y, color="tab:orange", label=label, linewidth=2)
```

```python
sample_size = 300
interval = 1 # 階級幅

bins = calc_bins(interval)

fig = plt.figure(figsize=(16, 4))
fig.suptitle(f"標本数: {sample_size}")

iteration = 3

for i in range(iteration):
    # ref: https://qiita.com/sakabe/items/df7fac64e6ffbb50848c
    data = generate_data(sample_size)
    
    ax = fig.add_subplot(1, iteration, i + 1)
    draw_histogram(data, ax, bins)
    draw_graph(ax, interval)
    ax.legend()

plt.savefig(f"sampling-height_{sample_size}.png", dpi=300, bbox_inches="tight")
```
